```python
# rename column name
file1 = file1.rename(columns={'ST':'State'})

file2 = file2.rename(columns={'ST':'State'})

file1.head()
```

```python
cols1 = []
for column in file1.columns:
    cols1.append(column.lower())
file1.columns = cols1
```

```python
cols2 = []
for column in file2.columns:
    cols2.append(column.lower())
file2.columns = cols2
```

```python
cols3 = []
for column in file3.columns:
    cols3.append(column.lower())
file3.columns = cols3
```

### Rearrange the columns in the dataframe as needed

```python
file1.columns
```

```python
file2 = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

file2.head()
```

```python
file3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

file3.head()
```

### Concatenate the three dataframes

```python
data = pd.concat([file1, file2, file3], axis=0)
data.shape
```

### Which columns are numerical?

```python
data.dtypes
```

```python
data._get_numeric_data().head()
```

### Which columns are categorical?

Categorical columns are: "state", "gender", "policy type", "vehicle class"

### Understand the meaning of all columns

### Perform the data cleaning operations

```python
data = data.drop(['education','number of open complaints'], axis=1)
data.head()
```

### Correct the values in the column customer lifetime value

```python
data['customer lifetime value'] = data['customer lifetime value'].str.replace('%','')
data.head()
```

```python
data['customer lifetime value'] =  pd.to_numeric(data['customer lifetime value'], errors='coerce')
data.dtypes
```

```python
data['customer lifetime value'] = list(map(lambda x: x/10000, data['customer lifetime value']))
data.head()
```

### Check for duplicate rows in the data and remove if any.

```python
display(data.shape)
display(data.duplicated().sum())
data = data.drop_duplicates()
display(data.shape)
```

### Filter out the data for customers who have an income of 0 or less

```python
data_zero = data[data["income"] <= 0]
display(data_zero.head())
```
